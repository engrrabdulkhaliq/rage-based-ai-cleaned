```python
import os
import json
import math

n = 5  

for filename in os.listdir("jsons"):
    if filename.endswith(".json"):
        file_path = os.path.join("jsons", filename)
        
        with open(file_path, "r", encoding="utf-8") as f:
            data = json.load(f)
        
        new_chunks = []
        num_chunks = len(data['chunks'])
        num_groups = math.ceil(num_chunks/n)
        
        for i in range(num_groups):
            start_idx = i*n
            end_idx = min((i+1)*n, num_chunks)
            
            chunk_group = data['chunks'][start_idx:end_idx]
            
            
            merged_text = " ".join([c['text'] for c in chunk_group])
        
            
            new_chunk = {
                'number': chunk_group[0]['number'],
                'title': chunk_group[0]['title'],
                'start': chunk_group[0]['start'],
                'end': chunk_group[-1]['end'],
                'text': merged_text
            }
            
            new_chunks.append(new_chunk)
        
        data['chunks'] = new_chunks
        
        os.makedirs("newjsons", exist_ok=True)
        output_path = os.path.join("newjsons", filename)
        
        with open(output_path, "w", encoding="utf-8") as f:
            json.dump(data, f, ensure_ascii=False, indent=4)
        
        print(f"‚úì Processed: {filename} ({num_chunks} ‚Üí {len(new_chunks)} chunks)")

print("\n‚úÖ All files merged successfully!")
print("üìÅ Output folder: newjsons/")
```
